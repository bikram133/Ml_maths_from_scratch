```python
#Normalization of min-max and mean-standardDeviation
import pandas as pd
import numpy as np
```

```python
def min_max(a):
  max=np.max(a)
  min=np.min(a)
  stand=[]
  for i in range(len(a)):
    stand.append((a[i]-min)/max-min)
  return stand
```

```python
def standarization(a):
  mean=np.mean(a)
  std=np.std(a)
  normalized=[]
  for i in range(len(a)):
    normalized.append((a[i]-mean)/std)
  return normalized
```
